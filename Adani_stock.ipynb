```python
import requests
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
url="https://equitypandit.com/historical-data/ADANIENT"
response=requests.get(url)
```

```python
response
```

```python
soup=BeautifulSoup(response.content,"lxml")
```

```python
table=soup.find("table")
```

```python
title=soup.find("title")
title
```

```python
adani=soup.find("tbody")
adani
```

```python
for i in adani.find_all("tr"):
    print(i)
```

```python
Date=[]
for i in adani.find_all("tr")[0::1]:
    Da=i.find_all("td")[0]
    Date.append(Da.text)
Date
```

```python
Price=[]
for i in adani.find_all("tr")[0::1]:
    Pr=i.find_all("td")[1]
    Price.append(Pr.text)
Price
```

```python
Open=[]
for i in adani.find_all("tr")[0::1]:
    Op=i.find_all("td")[2]
    Open.append(Op.text)
Open
```

```python
High=[]
for i in adani.find_all("tr")[0::1]:
    Hi=i.find_all("td")[3]
    High.append(Hi.text)
High
```

```python
Low=[]
for i in adani.find_all("tr")[0::1]:
    Lo=i.find_all("td")[4]
    Low.append(Lo.text)
Low
```

```python
Volume=[]
for i in adani.find_all("tr")[0::1]:
    Vol=i.find_all("td")[5]
    Volume.append(Vol.text)
Volume
```

```python
Change=[]
for i in adani.find_all("tr")[0::1]:
    Cha=i.find_all("td")[6]
    Change.append(Cha.text)
Change
```

```python
data={"Date":Date, "Price":Price ,"Open":Open ,"High":High ,"Low":Low ,"Volume": Volume ,"Change":Change}
```
